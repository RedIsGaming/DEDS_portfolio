Geef een overzicht met daarin alle unieke eerste adressen van klantvestigingen en de voornaam en achternaam van de verkopers die ‘Branch Manager’ zijn en aan deze vestigingen hebben verkocht (3 kolommen, 1 rij) 

```python
df = retailer_site.merge(sales_staff, left_on="RETAILER_CODE", how="inner", right_on="SALES_BRANCH_CODE", indicator=True)
subsets: list[str] = ["ADDRESS1", "FIRST_NAME", "LAST_NAME"]
data = df["POSITION_EN"].isin(["Branch Manager"])
df.loc[data, subsets].drop_duplicates(subset=df[subsets]).head(1)
```

Geef een overzicht met daarin van de verkopers hun functie en indien zij iets hebben verkocht de datum waarop de verkoop heeft plaatsgevonden. Laat alleen de verschillende namen van de posities zien van de verkopers die het woord ‘Manager’ in hun positienaam hebben staan. (2 kolommen, 9 rijen) 

```python
df = sales_staff.merge(order_header, on="SALES_STAFF_CODE", how="left", indicator=True)
data = df["POSITION_EN"].str.contains("Manager")
df.loc[data, ["POSITION_EN", "ORDER_DATE"]].drop_duplicates().fillna("Niks verkocht")
```

Geef een overzicht met daarin de verschillende namen van producten en bijbehorende namen van producttypen van de producten waarvoor ooit meer dan 750 stuks tegelijk verkocht zijn. (2 kolommen, 9 rijen) 

```python
df = product.merge(product_type, on="PRODUCT_TYPE_CODE", how="inner").merge(order_details, on="PRODUCT_NUMBER", how="inner", indicator=True)
data = df["QUANTITY"].apply(lambda x: x >= 750)
df.loc[data, ["PRODUCT_NAME", "PRODUCT_TYPE_EN"]].drop_duplicates(subset=["PRODUCT_NAME"])
```

Geef een overzicht met daarin de productnamen waarvan ooit meer dan 40% korting is verleend. De formule voor korting is: (unit_price - unit_sale_price) / unit_price (1 kolom, 8 rijen) 

```python
df = product.merge(order_details, on="PRODUCT_NUMBER", how="inner", indicator=True)
df["UNIT_PRICE"] = pd.to_numeric(df["UNIT_PRICE"])
df["UNIT_SALE_PRICE"] = pd.to_numeric(df["UNIT_SALE_PRICE"])
data = df.apply(lambda x: (x["UNIT_PRICE"] - x["UNIT_SALE_PRICE"]) / x["UNIT_PRICE"] > 0.4, axis=1)
df.loc[data, ["PRODUCT_NAME"]].drop_duplicates(subset=["PRODUCT_NAME"])
```

Geef een overzicht met daarin de retourreden van producten waarvan ooit meer dan 90% van de aangeschafte hoeveelheid is teruggebracht (return_quantity/quantity). (1 kolom, 3 rijen) 

```python
df = return_reason.merge(returned_item, on="RETURN_REASON_CODE", how="inner").merge(order_details, on="ORDER_DETAIL_CODE", how="inner", indicator=True)
df["RETURN_QUANTITY"] = pd.to_numeric(df["RETURN_QUANTITY"])
data = df.apply(lambda x: (x["RETURN_QUANTITY"] / x["QUANTITY"]) > 0.9, axis=1)
df.loc[data, ["RETURN_DESCRIPTION_EN"]].drop_duplicates(subset=["RETURN_DESCRIPTION_EN"])
```

Geef een overzicht met daarin van de producten behorend tot het producttype ‘Cooking Gear’ per productnaam de totaal verkochte hoeveelheid en de gemiddelde verkoopprijs. Sorteer de uitkomst op totaal verkochte hoeveelheid. (4 kolommen, 10 rijen) 

```python
df = product.merge(product_type, on="PRODUCT_TYPE_CODE", how="inner").merge(order_details, on="PRODUCT_NUMBER", how="inner", indicator=True)
df["UNIT_PRICE"] = pd.to_numeric(df["UNIT_PRICE"])
data = df["PRODUCT_TYPE_EN"].isin(["Cooking Gear"])
df = df.loc[data].groupby(["PRODUCT_TYPE_EN", "PRODUCT_NAME"], as_index=False).agg({"QUANTITY": "sum", "UNIT_PRICE": "mean"}).round(2).sort_values(by="QUANTITY", ascending=False)
df.index = range(1, len(df) + 1)
df
```

Geef een overzicht met daarin van de verkopers behorend tot de verkoopafdeling uit ‘Amsterdam’ de voornaam en achternaam van de verkoper en zijn laagste en hoogste verkooptarget van 2006. (4 kolommen, 4 rijen) 

```python
df = sales_branch.merge(sales_staff, on="SALES_BRANCH_CODE", how="inner").merge(target, on="SALES_STAFF_CODE", how="inner", indicator=True)
data = df.apply(lambda x: x["CITY"] == "Amsterdam" and x["SALES_YEAR"] == "2022", axis=1)
df = df.loc[data].groupby(["FIRST_NAME", "LAST_NAME"], as_index=False).agg({"SALES_TARGET": ["min", "max"]})
df.columns = [" ".join(columns) for columns in df.columns.values]
df.rename(columns={"SALES_TARGET min": "Laagste verkooptarget", "SALES_TARGET max": "Hoogste verkooptarget"}, inplace=True)
df.index = range(1, len(df) + 1)
df
```

Geef een overzicht met daarin per land de naam van het land, de naam van de stad waar de verkoopafdeling is gevestigd (noem de kolomnaam in het overzicht ‘verkoper’) en het aantal steden waar zich klanten bevinden in dat land (noem de kolomnaam in het overzicht ‘klanten’) (3 kolommen, 29 rijen) 

```python
df = country.merge(sales_branch, on="COUNTRY_CODE", how="inner", indicator=True)
df = df.groupby(["COUNTRY"], as_index=False).agg({"CITY": ["unique", "size"]})
df.columns = [" ".join(columns) for columns in df.columns.values]
df.rename(columns={"CITY unique": "Verkoper", "CITY size": "Klanten"}, inplace=True)
df = df.explode("Verkoper")
df["Verkoper"].apply(lambda x: ", ".join(map(str, x)))
df.index = range(1, len(df) + 1)
df
```
